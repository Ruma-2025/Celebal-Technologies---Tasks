```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#We are plotting a normal distribution for the height of men with data generated using mean height of 50 cm and a standard deviation of 5 cm.

mu, sigma = 150, 5
data = np.random.normal(mu, sigma, 10000)
```

```python
#Plotting the z-scores

sns.histplot(data, bins=30, kde=True, stat='probability')
plt.xlabel('Height (Men)')
plt.ylabel('Probability')
plt.xticks(range(155,200,5))
plt.title("Normal Distribution of Men's Height")
plt.show()
```

```python
#Calculating the % of men that are shorter than 140 cm

from scipy.stats import norm

z = (140-mu)/sigma
p = norm.cdf(z)

print(f"Percentage of men shorter than 183 cm is: {round(p*100,2)}%")
```
